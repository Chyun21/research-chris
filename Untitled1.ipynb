```python
# Run this cell to set up the notebook, but please don't change it.
import numpy as np
import math
from datascience import *

# These lines set up the plotting functionality and formatting.
import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

```python
vg = Table().read_table('vgsales.csv')
```

```python
vg_after_2010 = vg.where('Year', are.above(2010))
```

```python
vg_after_2010.sort('Year').group('Platform') 
```

```python
x_data = vg.sort("Year").column('Year')
y_data = vg.sort("Year").column('Global_Sales')

y_data2 = vg.sort("Year").column("JP_Sales")

y_data3 = vg.sort("Year").column('NA_Sales')

videogames = vg.group("Publisher")
count = videogames.where('count', are.above(200)).column('count')
names = videogames.where('count', are.above(200)).column('Publisher')

```

```python
plots.plot(x_data, y_data, label = "Global Sales")
plots.plot(x_data,y_data3, label = "NA Sales", linestyle = "--", linewidth=2)

plots.plot(x_data,y_data2, label = "Japan Sales")
plots.legend()
```

```python
vg.sort("Global_Sales", True)
```

```python
plots.pie(count, labels=names, autopct='%1.1f%%',shadow=True)
plots.show
```
